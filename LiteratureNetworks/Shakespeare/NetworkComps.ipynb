```python
from chAs import draw_comms_random,modul_arity
import community as comm

part=comm.best_partition(G) 
print 'Number of communities =', max(part.values())+1
print 'Community partition:'
parLis=[]
partdi={}
for i,k in part.items():
    if k not in partdi:
        partdi[k]=[i]
    else:
        partdi[k].append(i)
for i,k in partdi.items():
    parLis.append(k)
for nd in G.nodes():
    G.add_node(nd,attr_dict=G.node[nd],community=part[nd])   
print parLis
print 'Community modularity =', "%.4f" % comm.modularity(part,G)
print 'Community assortativity coefficient =', "%.4f" %nx.attribute_assortativity_coefficient(G,'community')

d=0.8 
dd=0.8
c=1.2
cc=1.4
alpha=0.2
ealpha=.2
vcc={}
sstt="The %s Communities" %(max(part.values())+1)

draw_comms_random(G,G.nodes(),[],[],[] ,part,part,d,dd,c,cc,alpha,ealpha,sstt,titlefont=18)
```
